```python
import csv

# Id, ParentId, PostTypeId, AcceptedAnswerId, Score, Body, CreationDate
data = {
    'id': [],
    'parentId': [],
    'postTypeId': [],
    'acceptedAnswerId': [],
    'score': [],
    'body': [],
    'creationDate': [],
}

with open('data.csv', 'r') as f:
    reader = csv.reader(f)
    header = next(reader)  # ヘッダーを読み飛ばしたい時

    cnt = 0
    for row in reader:
        data['id'].append(row[0])
        data['parentId'].append(row[1])
        data['postTypeId'].append(row[2])
        data['acceptedAnswerId'].append(row[3])
        data['score'].append(row[4])
        data['body'].append(row[5])
        data['creationDate'].append(row[6])
        
print(len(data['id']))
print(len(data['parentId']))
print(len(data['postTypeId']))
print(len(data['acceptedAnswerId']))
print(len(data['score']))
print(len(data['body']))
print(len(data['creationDate']))
```

```python
#　5.4 最初の分類機を作成する

# 最近傍法はモデルベースの手法でないため、どのようなデータにもそのまま学習させることが出来る
# しかし、これが欠点になることもある
```

```python
# 5.4.1 k近謗法からスタートする

# sklearn にある k近謗法ライブラリを利用する

from sklearn import neighbors
knn = neighbors.KNeighborsClassifier(n_neighbors=2)
print(knn)
```

```python
train_data = [[1], [2], [3], [4], [5], [6]]
train_label = [0, 0, 0, 1, 1, 1]
# 訓練を行う
knn.fit(train_data, train_label)
# 新しいラベルを推測
print(knn.predict(1.5))
print(knn.predict(100))
print(knn.predict(3))
```

```python
# 入力数値がどのクラスタに属するか確率を返す
print(knn.predict_proba(1.5))
print(knn.predict_proba(100))
print(knn.predict_proba(3.5))
```

```python
# 5.4.2 特徴量について模索

# 各回答文書回答から URL をカウントする

import re

code_match = re.compile('<pre>(.*?)</pre>', re.MULTILINE|re.DOTALL)
link_match = re.compile('<a href="http://.*?".*?>(.*?)</a>', re.MULTILINE|re.DOTALL)

def extract_feature_form_body(s):
    link_count_in_code = 0
    for match_str in code_match.findall(s):
        link_count_in_code += len(link_match.findall(match_str))
    return len(link_match.findall(s)) - link_count_in_code
```

```python
data['link_count'] = []
for elm in data['body']:
    count = extract_feature_form_body(elm)
    data['link_count'].append(count)
print(data['link_count'])
```
